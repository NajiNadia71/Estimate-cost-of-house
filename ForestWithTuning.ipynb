```python
DataSetAlgorithm=New.filter(['MSZoning','LotFrontage',
'LotShape','Neighborhood','Condition1',
'Condition2','OverallQual', 'MasVnrArea','ExterCond'
                               ,'SalePrice'],axis=1)
encoded_df =  pd.get_dummies(DataSetAlgorithm)
encoded_df.head()
```

```python
y = encoded_df["SalePrice"]
del encoded_df["SalePrice"]
encoded_df.dtypes
X = encoded_df
X.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

```python
n_estimators = [5,20,50,100] # number of trees in the random forest
max_features = ['auto', 'sqrt'] # number of features in consideration at every split
max_depth = [int(x) for x in np.linspace(10, 120, num = 12)] # maximum number of levels allowed in each decision tree
min_samples_split = [2,3,4, 6, 10] # minimum sample number to split a node
min_samples_leaf = [1, 3, 4] # minimum sample number that can be stored in a leaf node
bootstrap = [True, False] # method used to sample data points

random_grid = {'n_estimators': n_estimators,

'max_features': max_features,

'max_depth': max_depth,

'min_samples_split': min_samples_split,

'min_samples_leaf': min_samples_leaf,

'bootstrap': bootstrap}
```

```python
## Importing Random Forest Classifier from the sklearn.ensemble
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
from sklearn.model_selection import RandomizedSearchCV
rf_random = RandomizedSearchCV(estimator = rf,param_distributions = random_grid,
               n_iter = 100, cv = 5, verbose=2, random_state=35, n_jobs = -1)
rf_random.fit(X_train, y_train)
```

```python
randmf = RandomForestRegressor(n_estimators = 100, min_samples_split = 6, min_samples_leaf= 1, max_features = 'sqrt', max_depth= 100 , bootstrap=True) 
randmf.fit( X_train, y_train) 
```

```python
y_pred_rf1 = pd.DataFrame( { "actual": y_test, "predicted_prob": randmf.predict( ( X_test ) ) } ) 
```

```python
Y_Preb = randmf.predict( ( X_test ) ) 
```

```python
import sklearn.metrics as metrics
print(metrics.r2_score(y_test,Y_Preb))
```
